find an efficient path from its current location S to a desired location G. As an
idealization of the situation, assume that the obstacles (whatever they may be)
are abstracted by polygons. The problem now reduces to finding the shortest
path between two points in a plane that has convex polygonal obstacles.

```python
# defining the points in the problem

polygons = [
    [(2.0,0.5), (12.0, 0.5), (12.0, 4.0), (2.0, 4.0)],
    [(13.5, 2.0), (15.7, 4.5), (13.0, 7.2)],
    [(19.0, 0.7), (21, 1.8), (21.0, 4.0), (19.0,5.8), (17.0, 4.0), (17.0, 1.8)],
    [(5.0, 6.0), (7.0, 10.6), (5.0, 13.8), (1.0, 10.4), (1.8, 7.0)],
    [(7.0, 5.8), (10.0, 5.8), (8.5, 11.0)],
    [(11.0, 9.2), (14.0, 11.8), (13.0, 13.5), (11.3, 13.0)],
    [(14.8, 6.3), (18.7, 6.3), (18.7, 13.0), (14.8, 13.0)],
    [(21.4, 5.5), (22.4, 12.0), (21.0, 13.0), (20.0, 12.3)]
]
initial = (1.0, 2.0)
goal = (23.0, 13.3)
pol_list = [sg.Polygon(p) for p in polygons]

rb = RobotSearch(initial=initial, goal=goal, polygons=polygons)    
```

```python
def plot_polygons(pol_list, rb, title="Given Polygons with start and goal state", edges=False):
    '''A function to plot the given polygons with the start and goal state. The edges flag toggles displaying the
    edges between the vertices.'''
    new_shape = so.cascaded_union(pol_list)
    fig, axs = plt.subplots()
    axs.set_aspect('equal', 'datalim')

    for geom in new_shape.geoms:    
        xs, ys = geom.exterior.xy    
        axs.fill(xs, ys, alpha=0.5, fc='r', ec='none')

    if edges:
        for p1 in range(len(rb.adj)):
            for p2 in range(p1 + 1, len(rb.adj)):
                if rb.adj[p1][p2] != -1:
                    axs.plot([rb.p_list[p1][0], rb.p_list[p2][0]], [rb.p_list[p1][1], rb.p_list[p2][1]])    

    plt.scatter(*initial, label="Start State", c='green', s=65)
    plt.scatter(*goal, label="Goal State", c='red', s=65)
    plt.title(title)
    plt.legend()
    plt.show()
    
def plot_path(pol_list, rb, path, title="Given Polygons with start and goal state"):
    '''A function to plot the given polygons with the start and goal state and a path which is a set of points
    on the plane.'''
    new_shape = so.cascaded_union(pol_list)
    fig, axs = plt.subplots()
    axs.set_aspect('equal', 'datalim')

    for geom in new_shape.geoms:    
        xs, ys = geom.exterior.xy    
        axs.fill(xs, ys, alpha=0.5, fc='r', ec='none')
    
    for ind in range(len(path)):
        if ind != len(path)-1:
            axs.plot([rb.p_list[path[ind]][0], rb.p_list[path[ind+1]][0]], [rb.p_list[path[ind]][1], rb.p_list[path[ind+1]][1]])    
        axs.scatter([rb.p_list[path[ind]][0]], [rb.p_list[path[ind]][1]], s=50, c="blue")
    
    plt.scatter(*initial, label="Start State", c='green', s=65)
    plt.scatter(*goal, label="Goal State", c='red', s=65)
    plt.title(title)
    plt.legend()
    plt.show()    
```

```python
plot_polygons(pol_list, rb)
```

### How do we discretize this state space?

* The idea is to consider vertex to vertex hops alone, this ensures the state space is simplified and gives us the optimal path as well.
* The graph created using the vertices is called the visibility graph, shown below.

```python
plot_polygons(pol_list, rb, title="Given Polygons with Visibility Graph", edges=True)
```

## Testing Random Polygons:

For simplicity, I'm using 3 sided polygons to avoid issues with non-convex polygons being generated.

```python
def gen_polygons(n, initial, goal):    
    X_MIN = initial[0]
    X_MAX = goal[0]
    Y_MIN = 0.5
    Y_MAX = goal[1] + 1
    pols = []
    for i in range(n):
        inter = True
        while inter:
            pol = []
            num_points = 3  #random.randint(3, 5)
            for _ in range(num_points):
                pol.append((random.uniform(X_MIN, X_MAX), random.uniform(Y_MIN, Y_MAX)))
            flag = False
            for old_pol in pols:
                if sg.Polygon(pol).intersects(sg.Polygon(old_pol)):
                    flag = True
                    break
            if not flag:
                inter = False
                pols.append(sorted(pol))
    return pols

initial = (1.0, 2.0)
goal = (23.0, 13.3)
```

```python
for i in range(2):
    pols = gen_polygons(5, initial, goal)
    pol_list = [sg.Polygon(p) for p in pols]
    rb = RobotSearch(initial=initial, goal=goal, polygons=pols)    
    plot_polygons(pol_list, rb, edges=True)
```

```python
def test_algo(sizes, func):
    times = []
    for n in sizes:
        pols = gen_polygons(n, initial, goal)
        pol_list = [sg.Polygon(p) for p in pols]
        rb = RobotSearch(initial=initial, goal=goal, polygons=pols) 
        start = time.time()
        astar_path, cost = func()
        end = time.time()
        times.append(end - start)
    return times


sizes = [5, 10, 20, 50]
times_bfs = test_algo(sizes, rb.breadth_first_search_v1)
times_gbfs = test_algo(sizes, rb.greedy_best_first_search)
times_a_star = test_algo(sizes, rb.a_star_search)
```

## Results:

* A* is clearly the slowest but optimal algorithm.
* GBFS is the fastest but not optimal.
* BFS is not optimal but faster than A*.

```python
plt.plot(sizes, times_bfs, label="BFS")
plt.plot(sizes, times_gbfs, label="GBFS")
plt.plot(sizes, times_a_star, label="A*")
plt.legend()

plt.title("Execution Times (in s):")
```
